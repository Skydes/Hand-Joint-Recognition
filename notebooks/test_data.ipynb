```python
import h5py
from pathlib import Path
import numpy as np

from utils import plot_imgs
from jointrecog.settings import DATA_PATH
```

```python
def conv(im):
    return np.flip(np.moveaxis(im, 0, -1), -1)
```

# Test set

```python
test = h5py.File(Path(DATA_PATH, 'testing.h5'), 'r')['test']
print(len(test['img'])), print(test['img'][0].shape), print(test['img'][0].dtype)
plot_imgs([conv(test['img'][i])/255. for i in range(5)])
```

# Training set

```python
train = h5py.File(Path(DATA_PATH, 'training.h5'), 'r')['train']
print(len(train['img'])), print(train['img'][0].shape), print(train['img'][0].dtype)
plot_imgs([conv(train['img'][i])/255. for i in range(4)])
```
